```python
df2['Population'] = [x.replace(',','') for x in df2['Population']]
df2['Users'] = [x.replace(',','') for x in df2['Users']]
df2['Population'] = df2['Population'].astype(float)
df2['Users'] = df2['Users'].astype(float)
```

```python
from datetime import datetime, timedelta
import pandas as pd

def to_date_format(n):
    date_str = (datetime(1899, 12, 30) + timedelta(days=n-1)).strftime("%d-%m-%Y")
    date_date = datetime.strptime(date_str, "%d-%m-%Y")
    return date_date



# Apply to_date_format function to 'Date of Travel' column
df1['Date of Travel'] = df1['Date of Travel'].apply(to_date_format)
print(df1.head())
```

```python
df1.info()
df2.info()
df3.info()
df4.info()

```

```python
merged_data = pd.merge(df1, df4, on='Transaction ID')
merged_data = pd.merge(merged_data, df3, on='Customer ID')
merged_data = pd.merge(merged_data, df2, on='City')
```

```python
merged_data.info()
```

```python
merged_data.head()
```

```python
round(merged_data.describe(include='all',datetime_is_numeric=True),2)
```

```python
start_date = '2016-01-31'
end_date = '2018-12-31'

# Filter the data based on the time period
filtered_data = merged_data[(merged_data['Date of Travel'] >= start_date) & (merged_data['Date of Travel'] <= end_date)]

# Group the filtered data by the company and calculate the sum of users for each company
company_users = filtered_data.groupby('Company')['Users'].sum()

# Find the company with the maximum sum of users
max_company = company_users.idxmax()

# Print the result
print("The company with the maximum cab users is:", max_company)
```

```python
# Calculate the margin for each transaction
merged_data['Margin'] = merged_data['Price Charged'] - merged_data['Cost of Trip']

# Calculate the total margin and number of customers
total_margin = merged_data['Margin'].sum()
num_customers = merged_data['Customer ID'].nunique()

# Calculate the average margin per customer
average_margin = total_margin / num_customers

# Print the result
print("The average margin per customer is:", average_margin)
```

```python
# Hypothesis 1: Customers of Pink Cab and Yellow Cab have different average trip distances
avg_distance_by_company = merged_data.groupby('Company')['KM Travelled'].mean()
print("Average trip distance by company:")
print(avg_distance_by_company)
```

```python
# Hypothesis 2: The price charged by Pink Cab and Yellow Cab differs significantly
import scipy.stats as stats
t_stat, p_value = stats.ttest_ind(merged_data[merged_data['Company'] == 'Pink Cab']['Price Charged'],
                                 merged_data[merged_data['Company'] == 'Yellow Cab']['Price Charged'])
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

```

```python
# Hypothesis 3: There is a correlation between the cost of the trip and the price charged by the cab companies
correlation = merged_data['Cost of Trip'].corr(merged_data['Price Charged'])
print("Correlation between cost of trip and price charged:", correlation)

```

```python
# Hypothesis 4: Male and female customers have different average trip distances
avg_distance_by_gender = merged_data.groupby('Gender')['KM Travelled'].mean()
print("Average trip distance by gender:")
print(avg_distance_by_gender)

```

```python
# Hypothesis 5: Customers from different age groups have varying income levels
avg_income_by_age = merged_data.groupby('Age')['Income (USD/Month)'].mean()
print("Average income by age:")
print(avg_income_by_age)

```

```python
# Hypothesis 6: The number of users is positively correlated with the population of the city
correlation = merged_data['Population'].corr(merged_data['Users'])
print("Correlation between population and number of users:", correlation)
```

```python
# Hypothesis 7: Payment mode (Cash or Card) affects the average price charged by cab companies
avg_price_by_payment_mode = merged_data.groupby('Payment_Mode')['Price Charged'].mean()
print("Average price charged by payment mode:")
print(avg_price_by_payment_mode)
```
